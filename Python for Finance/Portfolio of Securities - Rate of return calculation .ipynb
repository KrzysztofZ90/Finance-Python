```python
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import numpy as np
```

```python
# IMPORTING PG, MSFT, F, GE DATA FROM YAHOO FINANCE
ticker = ['PG','MSFT','F','GE']
mydata = pd.DataFrame()
for t in ticker:
    mydata[t] = wb.DataReader(t,data_source='yahoo',start='1995-1-1')['Adj Close']

 #t = wb.DataReader(ticker,data_source='yahoo',start='1995-1-1')['Adj Close']
```
